2. Funcion de densidad $g(x)$
$$g(x)=\frac{t(x)}{C}$$
$$C=\int_{0}^{1}t(x)dx$$
$$C= \int a\sin(\pi x) dx= \frac{-a}{\pi}\cos(\pi x)$$
$$=\int_{0}^{1}a\sin(\pi x)dx= [\frac{-a}{\pi}\cos(\pi)] - [\frac{-a}{\pi} \cos(0)]$$
$$ [\frac{-a}{\pi}\cos(-1)] - [\frac{-a}{\pi} \cos(1)]$$
$$C=\frac{2a}{\pi}$$
$$g(x)=\frac{a\sin(\pi x)}{\frac{2a}{\pi}}$$
$$g(x)=\frac{\sin(\pi x)}{\frac{2}{\pi}}$$

3. Distribucion de probabilidad acumulada $G(x)$
$$G(x)=\int_{0}^{x} g(x)dx$$
$$=\int \frac{sin(\pi x)}{2\pi}dx= -(\frac{1}{2})\cos(\pi x)+C$$
$$\int_{0}^{x} \frac{\sin(\pi x)}{2\pi}dx=[-(\frac{1}{2}\pi)\cos(\pi x)]-[-(\frac{1}{2}\pi)\cos(\pi(0))]$$
$$=[-(\frac{1}{2}\pi)\cos(\pi x)]-[-(\frac{1}{2}\pi)]$$
$$G(x)=[-(\frac{1}{2}\pi)\cos(\pi x)]+ \frac{1}{2\pi}$$

Despejar $x$
$$G(x)=U$$
$$[-(\frac{1}{2}\pi)\cos(\pi x)]+ \frac{1}{2\pi}=U$$
$$[-(\frac{1}{2}\pi)\cos(\pi x)]=U-\frac{1}{2\pi}$$
$$\cos(\pi x)= -2\pi(U-\frac{1}{2\pi}) $$
$$\pi x= \arccos[-2\pi(U-\frac{1}{2\pi})]$$
$$x= \frac{1}{\pi} \arccos[-2\pi (U-\frac{1}{2\pi})]$$

```python
def distribucion(N):
    return ((1/np.pi)*np.arccos(-2*np.pi*(np.random.random(N)-(1/(2*np.pi)))))
N=10**6
g= lambda g:((np.sin(np.pi*x)/(2/np.pi)))
x= np.arange(0,1,.01)
x1= distribucion(N)
plt.plot(x,g(x),label='g(x)')
plt.hist(x1,100,density=True)
plt.xlabel('valores aleatorios')
plt.ylabel('probabilidad')
plt.title('histograma funci√≥n creada')
plt.show()
```

Definimos la funci√≥n: $$g(x)=\frac{t(x)}{c}$$

$$c=\int_{0}^{1} a \sin (\pi x) dx$$

Siendo  $u = \pi x$, entonces $du = \pi dx$ 

$$\frac{1}{\pi} \int_{0}^{1} a \sin (u) du = \frac{a}{\pi} [-\cos(u)]_{0}^{\pi} = \frac{2a}{\pi} $$

Por lo tanto la funci√≥n de densidad es:

$$g(x) = \frac{t(x)}{\int_{0}^{1}t(x) dx} = \frac{a\sin(\pi x)}{\frac{2a}{\pi}} = \frac{\pi}{2}\sin(\pi x)$$

$$ \frac{1}{2}-\frac{\cos (\pi x)}{2\pi} = U $$

Despejamos X

$$\cos(\pi x )  = 1-2\pi U $$
$$x = \frac{1}{\pi}\arccos(1-2\pi U)$$

5. Genere 500 puntos aleatorios usando el m√©todo de aceptaci√≥n y rechazo y las funciones $f(x)$ y $t(x)$ para validar que todos los c√°lculos anteriores est√°n correctamente realizados. El resultado debe de ser como sigue:


```python
#C√≥digo de soluci√≥n estudiante 1
def Acep_rechazo(R2:'Variables distruidas U~U(0,1)',
                 R1:'Variables distribuidas como g(x)',
                 f:'funci√≥n objetivo a generar',
                 t:'funci√≥n que mayora a f'):
    f_x = f(R1)
    t_x = t(R1)
    condition = R2*t_x <=f_x
    [plt.plot(R1[i],R2[i]*t_x[i],'ob') if condition[i] else plt.plot(R1[i],R2[i]*t_x[i],'o') \
         for i in range(len(R1))]
                
    plt.show()
```

```python
N = 500 # n√∫mero de puntos a simular
# Como estoy tomando t(x) constante solo es necesario generar valores aleatorios U~(0,1)
R2 = np.random.rand(N)
R1 = np.random.rand(N)

x2 = np.arange(0,1,0.01)

plt.plot(x2,f(x2),'k')
plt.plot(x2,t(x2),'b')

Acep_rechazo(R2, R1, f, t)
```
